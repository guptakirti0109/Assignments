**Q1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.**

Ans - Data are mainly of two types: 

i) Qualitative Data - It is the information that cannot be counted, measured or easily expressed using numbers. It is collected from text, audio and images.

ii) Quantitative Data - It is the numerical data that can be counted, measured or easily expressed as numbers.It is collected from different resources such as text books, surveys, manually, etc.

Now, Quantitaive Data is of two types:

a) Discete Data - Data that can only take certain values is called discrete data or discrete values. This is data that can be counted and has a limited number of values. It usually comes in the form of whole numbers or integers. 

Ex- No.of cars parked in the parking, No. of children in a family, etc.

b) Continuous Data - Data that can take infinite values is called continuous data or continuous values. This is data that can be measured and has a unlimited number of values. It usually comes in the form of real numbers.

Ex- Height of students, temperature measure, etc.

Also, Qualitative Data is of two types:

a) Nominal Data - Nominal data is the data that is not ordered or ranked. 

Ex- Gender, Religion, Location, etc.

b) Ordinal Data - Ordinal data is simply what the name implies: data that is ordered or ranked. 

Ex- Rank of a student in class, Grades, Rating, etc.


**Scales of measurement**

There are four sacles of measurement, two for measuring quantitative data and two for measuring qualitative data:



a) Nominal Scale - A nominal scale of measurement is used for qualitative data. It does not give any numerical meaning to the data. Using the nominal scale of measurement, the data can be classified but cannot be added, subtracted, multiplied, or divided.

Ex- Eye color, Country of birth.

b) Ordinal Data - The ordinal scale of measurement groups the data into order or rank. It contains the property of nominal scale as well, which is to classify data variables into specific labels. And in addition to that, it organizes data into groups though it does not have any numerical value.

Ex- 1 = happy,2 = neutral and 3 = unhappy.

c) Interval Scale - The interval scale contains properties of nominal and ordered data, but the difference between data points can be quantified. This type of data shows both the order of the variables and the exact differences between the variables. They can be added to or subtracted from each other, but not multiplied or divided. 

Ex- 40 degrees is not 20 degrees multiplied by two.

d) Ratio Scale - Ratio scales of measurement include properties from all four scales of measurement. The data is nominal and defined by an identity, can be classified in order, contains intervals and can be broken down into exact value. Data in the ratio scale can be added, subtracted, divided and multiplied.
Ratio scales also differ from interval scales in that the scale has a true zero

Ex- Weight, height and distance are all examples of ratio variables..

Example- 

**Dataset: 3,4,4,6,7,8,12,13,15,16,17**

**The median of the dataset is the value directly in the middle, which turns out to be 8**

**Use case:** It is best to use the median when the distribution is either skewed or there are outliers present.

**c) Mode-** The mode represents the data that occurs most commonly in the given dataset.
It is mostly used for categorical data.

Example-

**Here,4 occurs maximum no. of times.That's why 4 is the mode of the given dataset.**

**Q4. What is a box plot, and what can it tell you about the distribution of data?**

Ans - The process to abstract a set of data that is estimated using an interval scale is called a box and whisker plot. It is also called just a box plot. These are mostly used for data interpretation. It is one of the types of graphical methods which displays the variation of the data in the dataset. 

A box plot is a chart that shows data from a five-number summary including one of the measures of central tendency.

Box or Whiskers plot is a method used for depicting groups of numerical data through their quartiles graphically. These may also have some lines extending from the boxes or whiskers which indicates the variability outside the lower and upper quartiles, hence the terms box-and-whisker plot and box-and-whisker diagram. Outliers can be indicated as individual points.



**Parts of Boxplot**

**Minimum:** The minimum value in the given dataset.

**First Quartile(Q1):** The first quartile is the 25th percentile value of the given dataset.

**Median(Q2):** The median is the middle value of the dataset, which divides the given dataset into two equal parts. It is considered as the second quartile.Also, it is the 50th percentile value of the given dataset. 

**Third Quartile(Q3):** The third quartile is the 75th percentile value of the given dataset.

**Maximum:** The maximum value in the given dataset.

There are two more terms which are used here-

**Interquartile Range (IQR):** The difference between the third quartile and first quartile is known as the interquartile range.

(i.e.) IQR = Q3-Q1

**Outliers:** The data that fall outside the lower fence and the upper fence are called outliers.dataset.

**Q6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?**

Ans - Skewness can be defined as a statistical measure that describes the lack of symmetry or asymmetry in the probability distribution of a dataset. It quantifies the degree to which the data deviates from a perfectly symmetrical distribution.

Skewness is a valuable statistical term because it provides insight into the shape and nature of a datasets distribution. 

There are two types of Skewness:

**a) Positive Skewness-** In a positively skewed distribution, the tail on the right side (the larger values) is longer than the tail on the left side (the smaller values). This means that the majority of data points are concentrated on the left side of the distribution, and there are some extreme values on the right side.In the case of a positively skewed dataset,

**Mean > Median > Mode**

**b)Negative Skewness-** In a negatively skewed distribution, the tail on the left side (the smaller values) is longer than the tail on the right side (the larger values). This implies that most of the data points are concentrated on the right side of the distribution, with a few extreme values on the left side. In the case of a negatively skewed dataset,

**Mean < Median < Mode**

When the mean, median and mode are equal, then the data points are evenly distributed around the central point which indicates a perfectly symmetrical distribution.



Skewness affect the interpretation of data in following ways.

a) It explains how the data is spread and provides valuable insights for accurate interpretation of statistical modeling.

b) It helps to identify the presence of outliers in the data. Very long tails or large values of the skewness coefficient indicates the presence of extreme values in the data

c) It can provide a guide on which data transformation techniques you can employ when processing your data.)

**Q7. What is the interquartile range (IQR), and how is it used to detect outliers?**

Ans - **Interquartile Range (IQR):** The difference between the third quartile and first quartile is known as the interquartile range.

(i.e.) IQR = Q3-Q1

where Q1 is the first quartile and Q3 is the third quartile.

In a Box-Whisker plot, we show Five-points summary i.e.,Minimum value, Q1, Q2(Median), Q3 and Maximum Value.

There are some lines extending from the boxes or whiskers which indicates the variability outside the lower and upper quartiles known as Lower Whisker and Upper Whisker, which can be calculated with the help of IQR.

**Lower whisker = Q1 - 1.5 * IQR**

**Upper whisker = Q3 + 1.5 * IQR**

After calculating lower and upper whsiker, we are able to find some points which lie outside the lower and upper whisker known as outliers.

Hence, in this way IQR is helpful in detecting outliers.

**Q8. Discuss the conditions under which the binomial distribution is used.**

Ans - Binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, either Success or Failure.

There are two parameters **n** and **p** used here in a binomial distribution. The variable n states the number of times the experiment runs and the variable p tells the probability of any one outcome. 

The conditions under which the binomial distribution is used are as follows:

a) The number of observations is fixed.

b) Each observation is independent.

c) Each observation represents one of two outcomes i.e., success or failure.

d) The probability of success is same for each outcome.

**Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).**

Ans - In a normal distribution, data is symmetrically distributed with no skew. When plotted on a graph, the data follows a bell shape, with most values clustering around a central region and tapering off as they go further away from the center.

The properties of normal distribution are as follows:

a) The mean, median and mode are exactly the same.

b) The distribution is symmetric about the mean- half the values fall below the mean and half above the mean.

c) The distribution can be described by two values - the mean and the standard deviation.
The mean is the location parameter while the standard deviation is the scale parameter.

**Empirical Rule-**

The empirical rule, or the 68-95-99.7 rule, tells you where most of your values lie in a normal distribution.

a) Around 68% of values are within 1 standard deviation from the mean.


b) Around 95% of values are within 2 standard deviation from the mean.


c) Around 99.7% of values are within 3 standard deviation from the mean.
