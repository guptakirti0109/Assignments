On the basis of these two random variables, there are two types of probability distributions.

The main difference between a probability mass function (PMF) and a probability density function (PDF) is that a PMF is used for discrete random variables, while a PDF is used for continuous random variables.

The probability of a discrete random variable is exactly equal to a specific value and the probability of a continuous random variable will fall within a specific range of values.

b) **Population variance-** The Z-test requires that the population variance is known whereas the t-test is used when the population variance is unknown.

**Q9.Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?**

**Task: Write Python code to calculate the z-score and p-value for the given data.**

**Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.**

```python
from scipy.stats import norm
import math

sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25
alpha = 0.05       #significance level

#calculating Z-score
z_score =(sample_mean - population_mean)/(std_dev / math.sqrt(sample_size))

#calculating p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

if p_value < alpha:
    solution = 'Reject the null hypothesis'
else:
    solution = 'Fail to reject the null hypothesis'

#output
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Solution: {solution}")
```

**Q10.Simulate a bionomial distribution with 10 trials and a probability of success of 0.6 using python.Generate 1000 samples and plot the distribution.What is the expected mean and variance?** 

**Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.**

**Objective: Understand the properties of a bionomial distribution and verify them through simulation.**

```python
import numpy as np
import matplotlib.pyplot as plt

#parameters for binomial distribution
n = 10
p = 0.6
no_of_samples = 1000

samples = np.random.binomial(n,p, no_of_samples)

expected_mean = n * p
expected_variance = n*p*(1-p)

sample_mean = np.mean(samples)
sample_variance = np.var(samples)

plt.figure(figsize = (8,5))
plt.hist(samples,bins = np.arange(n + 2) - 0.5, density=True, edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Output the results
print(f"Expected Mean: {expected_mean}")
print(f"Expected Variance: {expected_variance:.2f}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Variance: {sample_variance:.2f}")
```
