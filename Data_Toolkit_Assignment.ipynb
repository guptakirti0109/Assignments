Q3.Explain the following terms.
a) The difference in np.array, np.asarray and np.asanyarray.
b) The difference between Deep copy and shallow copy.

Ans- a) Difference in np.array, np.asarray and np.asanyarray is as follows:


1.   **np.array()** - The np.array() in Python is used to convert a list,tuple,etc. into a Numpy array.
**Syntax**: np.array(object,dtype=None)
Parameters:
*   Object: [array_like] Input data, in any form can be converted to an array.
*   dtype: [data-type,optional] By default, the datatype is inferred from the input data.



---




2. **np.asarray()** - The np.asarray() function is used when we want to convert the input to an array.Input can be lists,lists of tuples,tuples of tuples,tuples of lists and arrays.
**Syntax**: np.asarray(arr,dtype=None,order=None)
Parameters:


*   arr: [array_like] Input data, in any form can be converted to an array. This includes lists, lists of tuples, tuples, tuples of tuples, tuples of lists and ndarrays.
*   dtype: [data-type, optional] By default, the data-type is inferred from the input data.


*   order: Whether to use row-major (C-style) or column-major (Fortran-style) memory representation. Defaults to ‘C’.



---


3. **np.asanyarray()** - The np.asanyarray() function is used to convert the input to an ndarray, but pass ndarray subclasses through.
**Syntax**: np.asanyarray(a,dtype=None,order=None)
Parameters:


*   a: [array_like] Input data, in any form that can be converted to an array. This includes scalars, lists, lists of tuples, tuples, tuples of tuples, tuples of lists, and ndarrays.
*   dtype: [data-type,optional] By default, the data-type is inferred from the input data.


*   order: Whether to use row-major (C-style) or column-major (Fortran-style) memory representation. Defaults to 'C'.















b) Difference between Deep copy and shallow copy is as follows:


1.   Deep Copy - In deep copy, the changes made in one variable doesn't reflect in the another variable.In other words, it creates a new object that is completely independent of the original object.
The memory locations are different for original object and the new object.
2.   Shallow Copy - In shallow copy,the changes made in one variable reflects in the another variable.
It creates a new object that references the same memory locations as the original object.

Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:
a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'

```python
# Create a time-series dataset
dates = pd.date_range(start='2024-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=10, high=30, size=100)
humidity = np.random.uniform(low=30, high=90, size=100)

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plot Temperature and Humidity on the same plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Set the title
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

Q19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
import random

# Prepare random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(10, 100) for _ in categories]

# Create a data source
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Define a color map based on values
colors = factor_cmap('categories', palette=['#718dbf', '#e84d60', '#ddb7b1', '#c9d9d3', '#fddbc7'], factors=categories)

# Create a new plot
p = figure(x_range= categories, title="Random Categorical Bar Chart", x_axis_label='Categories', y_axis_label='Values',height=400, width=700, toolbar_location=None, tools="")

# Add bar renderer with color mapping
p.vbar(x='categories', top='values', width=0.8, source=source, legend_field="categories", line_color='white', fill_color=colors)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Customize legend
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

Q20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'.

```python
import plotly.graph_objects as go

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.rand(100) * 10  # Random values between 0 and 10

# Create a simple line plot
fig = go.Figure()

# Add the line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Line'))

# Set the title and labels
fig.update_layout( title='Simple Line Plot', xaxis_title='X Axis', yaxis_title='Y Axis')

# Display the plot
fig.show()
```

Q21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python
import plotly.graph_objects as go

# Generate random data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(10, 100, size=len(categories))

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=categories, values=values, hole=.3, textinfo='label+percent', hoverinfo='label+percent+value')])

# Set the title
fig.update_layout(title='Interactive Pie Chart')

# Display the chart
fig.show()
```
